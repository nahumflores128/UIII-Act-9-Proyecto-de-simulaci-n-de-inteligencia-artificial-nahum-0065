```python
# =========================================================
# SIMULADOR DE IA PARA CONDUCTOR INTELIGENTE
# CBTIS 128
# Carrera: Inteligencia Artificial
# Materia: Elabora Proyectos de Programación Lógica
# Docente: Eliseo Nava
# Alumno: Nahum Flores NL = 21
# =========================================================

print("==========================================")
print("   IA PARA CONDUCTOR INTELIGENTE 🚗")
print("==========================================")

# La lista inicia vacía
conductores = []

opcion = 0

while opcion != 5:

    print("\n========= MENÚ PRINCIPAL =========")
    print("1. Registrar conductor")
    print("2. Ver conductores")
    print("3. Analizar conductor con IA")
    print("4. Eliminar último conductor")
    print("5. Salir")

    opcion = int(input("Seleccione una opción: "))

    # -------------------------------------------------
    # OPCIÓN 1
    # -------------------------------------------------
    if opcion == 1:

        nombre = input("Nombre del conductor: ")
        velocidad = int(input("Velocidad promedio (km/h): "))
        cansancio = input("¿El conductor está cansado? (si/no): ")
        combustible = int(input("Nivel de combustible (%): "))

        # Guardar datos en una lista
        datos = [nombre, velocidad, cansancio, combustible]

        conductores.append(datos)

        print("\n✅ Conductor registrado correctamente.")

    # -------------------------------------------------
    # OPCIÓN 2
    # -------------------------------------------------
    elif opcion == 2:

        if conductores == []:
            print("\n⚠️ No hay conductores registrados.")
        else:
            print("\n===== LISTA DE CONDUCTORES =====")

            i = 0

            while i < len(conductores):

                print("\nConductor", i + 1)
                print("Nombre:", conductores[i][0])
                print("Velocidad:", conductores[i][1], "km/h")
                print("Cansancio:", conductores[i][2])
                print("Combustible:", conductores[i][3], "%")

                i = i + 1

    # -------------------------------------------------
    # OPCIÓN 3
    # -------------------------------------------------
    elif opcion == 3:

        if conductores == []:
            print("\n⚠️ Primero registra un conductor.")

        else:

            numero = int(input("Ingrese número de conductor: "))

            if numero > len(conductores) or numero <= 0:
                print("❌ Conductor no encontrado.")

            else:

                nombre = conductores[numero - 1][0]
                velocidad = conductores[numero - 1][1]
                cansancio = conductores[numero - 1][2]
                combustible = conductores[numero - 1][3]

                print("\n===== ANALISIS IA =====")
                print("Conductor:", nombre)

                # Inteligencia artificial con lógica condicional
                if velocidad > 120 and cansancio == "si":
                    print("🚨 ALERTA MÁXIMA")
                    print("La IA detectó conducción peligrosa.")
                    print("Recomendación: detener el vehículo.")

                elif velocidad > 100:
                    print("⚠️ ALERTA")
                    print("La IA detectó exceso de velocidad.")

                elif combustible < 20:
                    print("⛽ ALERTA")
                    print("La IA detectó combustible bajo.")

                elif cansancio == "si":
                    print("😴 ALERTA")
                    print("La IA detectó cansancio del conductor.")

                else:
                    print("✅ Todo en orden.")
                    print("Conducción segura detectada.")

    # -------------------------------------------------
    # OPCIÓN 4
    # -------------------------------------------------
    elif opcion == 4:

        if conductores == []:
            print("\n⚠️ No hay conductores para eliminar.")

        else:
            eliminado = conductores.pop()

            print("\n🗑️ Último conductor eliminado:")
            print(eliminado[0])

    # -------------------------------------------------
    # OPCIÓN 5
    # -------------------------------------------------
    elif opcion == 5:

        print("\n👋 Cerrando sistema...")
        print("Gracias por usar la IA para conductores.")

    # -------------------------------------------------
    # OPCIÓN INVÁLIDA
    # -------------------------------------------------
    else:
        print("\n❌ Opción incorrecta.")

print("\nPrograma creado por Nahum Flores NL = 21")
```
